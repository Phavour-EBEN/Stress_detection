```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from imblearn.over_sampling import SMOTE  # ✅ new
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('Updated_Psychological_Status.xlsx')

# Check for missing values
print(df.isnull().sum())

# Quick summary
print(df.info())
print(df.describe())

# Class distribution
print(df['STATUS'].value_counts())
print(df['STATUS'].unique())
```

```python
le = LabelEncoder()
df['STATUS'] = le.fit_transform(df['STATUS'])  # Now classes are 0,1,2,...
print(le.classes_)  # Map of encoded labels
```

```python
X = df.drop('STATUS', axis=1)
y = df['STATUS']

# Normalize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y  # ✅ stratify to keep balance
)
```

```python
from collections import Counter
print("Before SMOTE:", Counter(y_train))

sm = SMOTE(random_state=42)
X_train, y_train = sm.fit_resample(X_train, y_train)

print("After SMOTE:", Counter(y_train))
```

```python
rf_model = RandomForestClassifier(
    n_estimators=100,
    max_depth=4,              # Regularization: shallow trees to reduce overfitting
    min_samples_leaf=5,       # Avoid tiny splits
    class_weight='balanced',  # Helps with imbalanced classes
    random_state=42
)

rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_test)
print("Random Forest Report:\n", classification_report(y_test, y_pred_rf, zero_division=0))
```

```python
svm_model = SVC(kernel='rbf', C=1, gamma='scale', class_weight='balanced')
svm_model.fit(X_train, y_train)
y_pred_svm = svm_model.predict(X_test)

print("SVM Report:\n", classification_report(y_test, y_pred_svm, zero_division=0))
```

```python
def plot_confusion(y_true, y_pred, model_name):
    cm = confusion_matrix(y_true, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=le.classes_,
                yticklabels=le.classes_)
    plt.title(f'Confusion Matrix: {model_name}')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()

plot_confusion(y_test, y_pred_rf, "Random Forest")

# Train and predict with SVM model before plotting its confusion matrix
svm_model = SVC(kernel='rbf', C=1, gamma='scale', class_weight='balanced')
svm_model.fit(X_train, y_train)
y_pred_svm = svm_model.predict(X_test)

plot_confusion(y_test, y_pred_svm, "SVM")
```

```python
# Use the original column names from your DataFrame
input_df = pd.DataFrame([[80.5, 06.355, 8.08897, 80.054613, 7.03]],
                        columns=X.columns)

# Then transform using the scaler (no warning)
scaled_sample = scaler.transform(input_df)

# Predict
pred = rf_model.predict(scaled_sample)
print("Predicted Class:", le.classes_[pred[0]])
```

```python
from tensorflow.keras import Input
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

num_classes = len(y.unique()) # Define num_classes here

model = Sequential([
    Input(shape=(X_train.shape[1],)),
    Dense(64, activation='relu'),
    Dense(32, activation='relu'),
    Dense(num_classes, activation='softmax')
])
```

Finally, let's visualize the ROC curves for both the Random Forest and SVM models to assess their performance across different classification thresholds.

```python
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize

# Binarize the output for ROC curve analysis (multiclass)
y_test_bin = label_binarize(y_test, classes=np.unique(y))

# Get predicted probabilities (Random Forest)
y_prob_rf = rf_model.predict_proba(X_test)

# Get decision function scores (SVM)
y_score_svm = svm_model.decision_function(X_test)

# Compute ROC curve and ROC area for each class
fpr_rf = dict()
tpr_rf = dict()
roc_auc_rf = dict()

fpr_svm = dict()
tpr_svm = dict()
roc_auc_svm = dict()

n_classes = len(le.classes_)

for i in range(n_classes):
    fpr_rf[i], tpr_rf[i], _ = roc_curve(y_test_bin[:, i], y_prob_rf[:, i])
    roc_auc_rf[i] = auc(fpr_rf[i], tpr_rf[i])

    fpr_svm[i], tpr_svm[i], _ = roc_curve(y_test_bin[:, i], y_score_svm[:, i])
    roc_auc_svm[i] = auc(fpr_svm[i], tpr_svm[i])

# Plot ROC curves
plt.figure(figsize=(9, 5))

colors = ['blue', 'red', 'green', 'purple'] # Define colors for each class

for i, color in zip(range(n_classes), colors):
    plt.plot(fpr_rf[i], tpr_rf[i], color=color, lw=2,
             label='Random Forest ROC curve of class {0} (area = {1:0.2f})'.format(le.classes_[i], roc_auc_rf[i]))
    plt.plot(fpr_svm[i], tpr_svm[i], color=color, lw=2, linestyle='--',
             label='SVM ROC curve of class {0} (area = {1:0.2f})'.format(le.classes_[i], roc_auc_svm[i]))

plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve for Multiclass')
plt.legend(loc="lower right")
plt.show()
```
